### UKCP18 hourly
Data is already just summer

```python
np.nanmax(ukcp18_hrly_2001_2019)
```

```python
ems = ['01','04', '05', '06', '07', '08', '09', '10', '11', '12', '13', '15']

ukcp18_hrly_2001_2014 = np.array([])
ukcp18_hrly_2001_2019 = np.array([])
ukcp18_hrly_2006_2019 = np.array([])

for em in ems:
    print(em)
    datadir = f'/nfs/a319/gy17m2a/PhD/ProcessedData/TimeSeries/UKCP18_hourly/2.2km_bng_regridded_12km_masked/2002_2020/{em}/'
    for yr in range(2001,2020):
        # Read in data
        file_name = datadir + f"compressed_{yr}_GB_jja.npy"
        this_yr = np.load(file_name)
        print(yr, len(this_yr))
        
        # Full data
        ukcp18_hrly_2001_2019 = np.append(ukcp18_hrly_2001_2019, this_yr)
        
        # Just years overlapping with radar
        if yr >=2006:
            ukcp18_hrly_2006_2019 =np.append(ukcp18_hrly_2006_2019, this_yr)
            
        # Just years overlapping CEH-GEAR
        if yr <=2014:
            ukcp18_hrly_2001_2014 =np.append(ukcp18_hrly_2001_2014, this_yr)           
```

```python
np.save('/nfs/a319/gy17m2a/PhD/datadir/PDF_Plotting/ukcp18_hrly_2001_2014_jja_GB.npy', ukcp18_hrly_2001_2014)
np.save('/nfs/a319/gy17m2a/PhD/datadir/PDF_Plotting/ukcp18_hrly_2001_2019_jja_GB.npy', ukcp18_hrly_2001_2019)
np.save('/nfs/a319/gy17m2a/PhD/datadir/PDF_Plotting/ukcp18_hrly_2006_2019_jja_GB.npy', ukcp18_hrly_2006_2019)
```
