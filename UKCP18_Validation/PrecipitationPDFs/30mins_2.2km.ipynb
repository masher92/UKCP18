### UKCP18 30 mins

```python
ukcp18_30m_2001_2014 = []
ukcp18_30m_2001_2019 = []
ukcp18_30m_2006_2019 = []
resolution = "2.2km_bng_masked"
ems= ['bc012', 'bc005', 'bc006', 'bc007', 'bc009', 'bc010', 'bc011', 'bc013',  'bc015',  'bc016', 'bc017', 'bc018']
for em in ems:
    print(em)
    datadir = f"/nfs/a319/gy17m2a/PhD/ProcessedData/TimeSeries/UKCP18_every30mins/{resolution}/2002_2020/{em}_wholeyear/"
    for yr in range(2001, 2020):
        file_name = datadir + f"compressed_{yr}_GB_jja.npy"
        this_yr = np.load(file_name)
        this_yr = this_yr.astype('float32')
        print(em, yr, len(this_yr))
        
        # Filter out values below 0.1
        this_yr = this_yr[this_yr >= 0.1]
        print(em, yr, len(this_yr))

        # Append to lists
        ukcp18_30m_2001_2019.append(this_yr)

        if yr >= 2006:
            ukcp18_30m_2006_2019.append(this_yr)

        if yr <= 2014:
            ukcp18_30m_2001_2014.append(this_yr)

# Concatenate lists into arrays
ukcp18_30m_2001_2014 = np.concatenate(ukcp18_30m_2001_2014)
ukcp18_30m_2001_2019 = np.concatenate(ukcp18_30m_2001_2019)
ukcp18_30m_2006_2019 = np.concatenate(ukcp18_30m_2006_2019)

# Save the arrays
np.save(f'/nfs/a319/gy17m2a/PhD/datadir/PDF_Plotting/ukcp18_30m_2001_2014_{resolution}_jja_GB.npy', ukcp18_30m_2001_2014)
np.save(f'/nfs/a319/gy17m2a/PhD/datadir/PDF_Plotting/ukcp18_30m_2001_2019_{resolution}_jja_GB.npy', ukcp18_30m_2001_2019)
np.save(f'/nfs/a319/gy17m2a/PhD/datadir/PDF_Plotting/ukcp18_30m_2006_2019_{resolution}_jja_GB.npy', ukcp18_30m_2006_2019)
```

```python
ems= ['bc005', 'bc006', 'bc007', 'bc009', 'bc010', 'bc011', 'bc013',  'bc015',  'bc016', 'bc017', 'bc018', 'bc012']
resolutions = ["2.2km_bng_masked", ]
for resolution in resolutions:
    ukcp18_30m_2001_2014 = np.array([])
    ukcp18_30m_2001_2019 = np.array([])
    ukcp18_30m_2006_2019 = np.array([])

    for em in ems:
        print(em)
        datadir = f"/nfs/a319/gy17m2a/PhD/ProcessedData/TimeSeries/UKCP18_every30mins/{resolution}/2002_2020/{em}_wholeyear/"
        for yr in range(2001,2020):
            # Read in data
            file_name = datadir + f"{yr}_compressed_GB_jja.npy"
            this_yr = np.load(file_name)
            print(em, yr, len(this_yr))
            this_yr = this_yr[this_yr >= 0.1]
            print(em, yr, len(this_yr))

            # Full data
            ukcp18_30m_2001_2019 = np.append(ukcp18_30m_2001_2019, this_yr)

            # Just years overlapping with radar
            if yr >=2006:
                ukcp18_30m_2006_2019 =np.append(ukcp18_30m_2006_2019, this_yr)

            # Just years overlapping CEH-GEAR
            if yr <=2014:
                ukcp18_30m_2001_2014 =np.append(ukcp18_30m_2001_2014, this_yr)   
                
    np.save(f'/nfs/a319/gy17m2a/PhD/datadir/PDF_Plotting/ukcp18_30m_2001_2014_{resolution}_jja_GB.npy', ukcp18_30m_2001_2014)
    np.save(f'/nfs/a319/gy17m2a/PhD/datadir/PDF_Plotting/ukcp18_30m_2001_2019_{resolution}_jja_GB.npy', ukcp18_30m_2001_2019)
    np.save(f'/nfs/a319/gy17m2a/PhD/datadir/PDF_Plotting/ukcp18_30m_2006_2019_{resolution}_jja_GB.npy', ukcp18_30m_2006_2019)
```
