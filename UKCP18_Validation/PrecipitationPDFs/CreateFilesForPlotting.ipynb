### CEH-GEAR hourly
Data is already just summer

```python
cehgear_data_2001_2014 = []
cehgear_data_1990_2014 = []
resolution = "1km_reformatted"

print(resolution)
if resolution == '1km_reformatted':
    datadir = f'/nfs/a319/gy17m2a/PhD/ProcessedData/TimeSeries/CEH-GEAR/{resolution}/'
else:
    datadir = f'/nfs/a319/gy17m2a/PhD/ProcessedData/TimeSeries/CEH-GEAR/{resolution}/AreaWeighted/'        

for yr in range(1990, 2017):
    print(yr)
    print(len(cehgear_data_1990_2014))
    file_name = datadir + f"{yr}_compressed_GB_jja.npy"
    this_yr = np.load(file_name).astype('float32')
    
    print(np.nanmin(this_yr))
    this_yr = this_yr[this_yr >= 0.1]
    
    cehgear_data_1990_2014.extend(this_yr)  # Use extend() to add multiple elements
    if yr >= 2001:
        cehgear_data_2001_2014.extend(this_yr)

# Convert lists to NumPy arrays at the end
cehgear_data_1990_2014 = np.array(cehgear_data_1990_2014, dtype='float32')
cehgear_data_2001_2014 = np.array(cehgear_data_2001_2014, dtype='float32')

# Save the arrays
np.save(f'/nfs/a319/gy17m2a/PhD/ProcessedData/PDF_Plotting/CEH-GEAR/cehgear_2001-2014_{resolution}_GB_jja.npy', cehgear_data_2001_2014)
np.save(f'/nfs/a319/gy17m2a/PhD/ProcessedData/PDF_Plotting/CEH-GEAR/cehgear_1990-2014_{resolution}_GB_jja.npy', cehgear_data_1990_2014)
```

### UKCP18 hourly
Data is already just summer

```python
season='jja'
ems = ['07', '08', '09', '10', '11', '12', '13', '15', '01','04', '05', '06', ]
resolutions = ["2.2km_bng_masked"] #"2.2km_bng_masked" "2.2km_bng_regridded_12km_masked" 
for resolution in resolutions:
    print(resolution)
    ukcp18_hrly_2001_2014 = np.array([])
    ukcp18_hrly_2001_2019 = np.array([])
    ukcp18_hrly_2006_2019 = np.array([])

    for em in ems:
        print(em)
        datadir = f'/nfs/a319/gy17m2a/PhD/ProcessedData/TimeSeries/UKCP18_hourly/{resolution}/2002_2020/{em}_{season}/'
        for yr in range(2001,2020):
            # Read in data
            file_name = datadir + f"compressed_{yr}_GB_{season}.npy"
            this_yr = np.load(file_name)
            print(len(this_yr))
            this_yr = this_yr[this_yr >= 0.1]
            print(len(this_yr))

            # Full data
            ukcp18_hrly_2001_2019 = np.append(ukcp18_hrly_2001_2019, this_yr)

            # Just years overlapping with radar
            if yr >=2006:
                ukcp18_hrly_2006_2019 =np.append(ukcp18_hrly_2006_2019, this_yr)

            # Just years overlapping CEH-GEAR
            if yr <=2014:
                ukcp18_hrly_2001_2014 =np.append(ukcp18_hrly_2001_2014, this_yr)  
        print(yr, len(this_yr))
        
    np.save(f'/nfs/a319/gy17m2a/PhD/ProcessedData/PDF_Plotting/UKCP18_hourly/ukcp18_hrly_2001_2014_{resolution}_GB_{season}.npy', ukcp18_hrly_2001_2014)
    np.save(f'/nfs/a319/gy17m2a/PhD/ProcessedData/PDF_Plotting/UKCP18_hourly/ukcp18_hrly_2001_2019_{resolution}_GB_{season}.npy', ukcp18_hrly_2001_2019)
    np.save(f'/nfs/a319/gy17m2a/PhD/ProcessedData/PDF_Plotting/UKCP18_hourly/ukcp18_hrly_2006_2019_{resolution}_GB_{season}.npy', ukcp18_hrly_2006_2019)
```

### UKCP18 30 mins

```python
resolution = "2.2km_bng_regridded_12km_masked" #"2.2km_bng_masked"

ukcp18_30m_2001_2014 = []
ukcp18_30m_2001_2019 = []
ukcp18_30m_2006_2019 = []

ems= ['bc012', 'bc005', 'bc006', 'bc007', 'bc009', 'bc010', 'bc011', 'bc013',  'bc015',  'bc016', 'bc017', 'bc018']
for em in ems:
    print(em)
    datadir = f"/nfs/a319/gy17m2a/PhD/ProcessedData/TimeSeries/UKCP18_every30mins/{resolution}/2002_2020/{em}_wholeyear/"
    for yr in range(2001, 2020):
        file_name = datadir + f"compressed_{yr}_GB_jja.npy"
        this_yr = np.load(file_name)
        this_yr = this_yr.astype('float32')
        print(em, yr, len(this_yr))
        
        # Filter out values below 0.1
        this_yr = this_yr[this_yr >= 0.1]
        print(em, yr, len(this_yr))

        # Append to lists
        ukcp18_30m_2001_2019.append(this_yr)

        if yr >= 2006:
            ukcp18_30m_2006_2019.append(this_yr)

        if yr <= 2014:
            ukcp18_30m_2001_2014.append(this_yr)

# Concatenate lists into arrays
ukcp18_30m_2001_2014 = np.concatenate(ukcp18_30m_2001_2014)
ukcp18_30m_2001_2019 = np.concatenate(ukcp18_30m_2001_2019)
ukcp18_30m_2006_2019 = np.concatenate(ukcp18_30m_2006_2019)

np.save(f'/nfs/a319/gy17m2a/PhD/ProcessedData/PDF_Plotting/UKCP18_30mins/ukcp18_30m_2001_2014_{resolution}_jja_GB.npy', ukcp18_30m_2001_2014)
np.save(f'/nfs/a319/gy17m2a/PhD/ProcessedData/PDF_Plotting/UKCP18_30mins/UKCP18_30mins/ukcp18_30m_2001_2019_{resolution}_jja_GB.npy', ukcp18_30m_2001_2019)
np.save(f'/nfs/a319/gy17m2a/PhD/ProcessedData/PDF_Plotting/UKCP18_30mins/ukcp18_30m_2006_2019_{resolution}_jja_GB.npy', ukcp18_30m_2006_2019)
```
