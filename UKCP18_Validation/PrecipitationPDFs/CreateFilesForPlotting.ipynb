```python
cehgear = []
resolution = "2.2km_regridded"

output_file = f'/nfs/a319/gy17m2a/PhD/ProcessedData/PDF_Plotting/CEH-GEAR/1990_2016_{resolution}_GB_{season}.npy'
output_file_filtered = f'/nfs/a319/gy17m2a/PhD/ProcessedData/PDF_Plotting/CEH-GEAR/2001_2016_{resolution}_GB_{season}.npy'

print(resolution)
if resolution == '1km_reformatted':
    datadir = f'/nfs/a319/gy17m2a/PhD/ProcessedData/TimeSeries/CEH-GEAR/{resolution}/'
else:
    datadir = f'/nfs/a319/gy17m2a/PhD/ProcessedData/TimeSeries/CEH-GEAR/{resolution}/AreaWeighted/'        

for yr in range(1990,2017):
    file_name = os.path.join(datadir + f"{yr}_compressed_GB_{season}.npy")

    if os.path.exists(file_name):
        # Load yearly data
        this_yr = np.load(file_name, mmap_mode='r')  # Memory-map for efficiency
        print(f"Processing: {yr} | Original size: {this_yr.size}")

        # Filter out values below 0.1
        this_yr = this_yr[this_yr >= 0.1]
        print(f"Filtered size: {this_yr.size}")

        # Append to the output file incrementally
        with open(output_file, 'ab') as f:
            np.save(f, this_yr)
            
        if yr >=2001:
            # Append to the output file incrementally
            with open(output_file_filtered, 'ab') as f:
                np.save(f, this_yr)
```

### UKCP18 hourly
Data is already just summer

```python
import os
ems = ['01','04', '05', '06', '07', '08', '09', '10', '11', '12', '13', '15', ]
season = 'wholeyear'
resolution = "2.2km_bng_masked"

output_file = f'/nfs/a319/gy17m2a/PhD/ProcessedData/PDF_Plotting/UKCP18_hourly/2001_2019_{resolution}_GB_{season}.npy'

# Ensure the output file does not exist before starting
if os.path.exists(output_file):
    os.remove(output_file)

for em in ems:
    datadir = f'/nfs/a319/gy17m2a/PhD/ProcessedData/TimeSeries/UKCP18_hourly/{resolution}/2002_2020/{em}_{season}/'

    for yr in range(2001, 2020):
        file_name = os.path.join(datadir, f"compressed_{yr}_GB_{season}.npy")

        if os.path.exists(file_name):
            # Load yearly data
            this_yr = np.load(file_name, mmap_mode='r')  # Memory-map for efficiency
            print(f"Processing: {em} {yr} | Original size: {this_yr.size}")

            # Filter out values below 0.1
            this_yr = this_yr[this_yr >= 0.1]
            print(f"Filtered size: {this_yr.size}")

            # Append to the output file incrementally
            with open(output_file, 'ab') as f:
                np.save(f, this_yr)

print(f"Data saved to: {output_file}")
```
