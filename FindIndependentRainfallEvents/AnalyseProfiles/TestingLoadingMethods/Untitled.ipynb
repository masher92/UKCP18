```python
import numpy as np
import matplotlib.pyplot as plt
import sys

sys.path.insert(1, '../../ProcessEvents')
from ProcessEventsFunctions import *

# Define the rainfall data for the original and mirrored events
rainfall_original = [60, 5, 5, 10, 10, 5, 3, 2, 0, 0]
rainfall_original = [3,3,3,3,6]
rainfall_original = [1,1,2,1,4]
rainfall_mirrored = rainfall_original[::-1]  # Reverse the original event to get the mirrored event

def calculate_d50(rainfall):
    # Compute the cumulative rainfall
    cumulative_rainfall = np.cumsum(rainfall)
    total_rainfall = cumulative_rainfall[-1]

    # Calculate the D50 point (when cumulative rainfall reaches 50% of the total)
    d50_index = np.where(cumulative_rainfall >= 0.5 * total_rainfall)[0][0]  # Find the first index that reaches 50%
    d50_percentage = (d50_index + 1) / len(rainfall) * 100  # Convert to percentage of total duration

    return d50_percentage

time_steps = np.arange(1, len(rainfall_original) + 1)

# Calculate D50 for the original and mirrored events
d50_original = calculate_d50(rainfall_original)
d50_mirrored = calculate_d50(rainfall_mirrored)

d50_interp_original = calc_d50_with_interpolation(rainfall_original)
d50_interp_mirrored = calc_d50_with_interpolation(rainfall_mirrored)

# Display the results
print(f"D50 for Original Event: {d50_original:.2f}% of event duration")
print(f"D50 for Mirrored Event: {d50_mirrored:.2f}% of event duration")

print(f"D50 for Original Event: {d50_interp_original:.2f}% of event duration")
print(f"D50 for Mirrored Event: {d50_interp_mirrored:.2f}% of event duration")

print(f"reversed is {100-d50_interp_original:.2f}")

fig,axs=plt.subplots(ncols=2)
axs[0].bar(time_steps, rainfall_original)
axs[1].bar(time_steps, rainfall_mirrored)
```
