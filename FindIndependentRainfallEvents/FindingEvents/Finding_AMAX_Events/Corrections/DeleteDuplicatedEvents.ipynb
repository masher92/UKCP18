```python
for gauge_num in range(0, 1294):
    if gauge_num not in [444, 827, 888]:
        directory_path = f"/nfs/a161/gy17m2a/PhD/ProcessedData/IndependentEvents/UKCP18_30mins/{em}/{gauge_num}/WholeYear/"
        
        # List files with size 0
        files_with_size_0 = [name for name in os.listdir(directory_path)
                             if os.path.isfile(os.path.join(directory_path, name)) 
                             and 'part0' in name 
                             and os.path.getsize(os.path.join(directory_path, name)) == 0]
        
        # Check file count and files with size 0
        file_count = len([name for name in os.listdir(directory_path)
                          if os.path.isfile(os.path.join(directory_path, name)) 
                          and 'part0' in name])
        
        # Print a message if file count is not as expected
        if file_count == 133:
            pass
        else:
            print(f"Gauge number {gauge_num}: Not as expected, {file_count} files")
        
        # Print out files with size 0 and delete them
        if files_with_size_0:
            print(f"Gauge number {gauge_num}: Files with size 0 - {files_with_size_0}")
            for file_name in files_with_size_0:
                file_path = os.path.join(directory_path, file_name)
                try:
                    os.remove(file_path)
                    print(f"Deleted file: {file_path}")
                except Exception as e:
                    print(f"Error deleting file {file_path}: {e}")
```
